```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Data/petrol_consumption.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df.describe()
```

### Label and Features

```python
X = df.drop('Petrol_Consumption', axis=1)
y = df['Petrol_Consumption']

X.shape, y.shape
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

### Model Training

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()

model.fit(X_train, y_train)
```

### Making Predictions

```python
y_preds = model.predict(X_test)
y_preds
```

### Evaluating Predictions

```python
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay, mean_absolute_error, mean_squared_error
```

```python
accuracy = accuracy_score(y_test, y_preds)
print(f'Accuracy of the Model: {round((accuracy*100),2)}%')
```

```python
print('Mean Absolute Error:', mean_absolute_error(y_test, y_preds))
print('Mean Squared Error:', mean_squared_error(y_test, y_preds))
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test, y_preds)))
```
