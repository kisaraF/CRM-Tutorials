### About the DataFrame

1.	`ReceiptNo` : Invoice number. Nominal, a 6-digit integral number that uniquely assigned to each transaction. Any number that starts with letter ‘c’ indicates a cancellation.
2.	`ProductCode` : Product code. Nominal, a 5-digit integral number uniquely assigned to each distinct product.
3.	`Description` : Product name. Nominal.
4.	`Quantity` : The quantities of each product (item) per transaction. Numeric.
5.	`ReceiptDate` : Invoice Date and time. Numeric, the day and time when each transaction was generated.
6.	`UnitPrice` : Numeric, product price per unit in sterling.
7.	`CustomerID` : Nominal, a 5-digit integral number uniquely assigned to each customer.
8.	`Country` : Nominal, the name of the country where each customer resides

```python
#Checking if the database is created and data was inserted through running a query
df_q = pd.read_sql('SELECT * FROM transactions', con)
df_q
```

Since the data includes Cancelled transactions, NULLs in CustomerIDs, 0 in UnitPrice they must be omitted.

Also We need to get the data in a group by applied to Customer ID to see Last Order Date for Recency, Count of Orders for Frequency and Sum of Unit Price for Monetary

```python
query = '''
SELECT CustomerID, MAX(ReceiptDate) AS LastOrderDate, COUNT(*) AS OrderCount, SUM(UnitPrice * Quantity) AS TotalPrice
FROM transactions 
WHERE ReceiptNo NOT LIKE 'C%' 
AND CustomerID IS NOT NULL 
AND UnitPrice <> 0
GROUP BY CustomerID
'''
```

```python
clean_data = pd.read_sql(query, con)
clean_data
```

```python
#WE need to write this clean data into a table in the same database
clean_data.to_sql('clean_data', con, if_exists = 'replace', index = False)
```

```python
query = '''
SELECT CustomerID, RFM_Recency, RFM_Frequency, RFM_Monetary, RFM_Recency * 100 + RFM_Frequency * 10 + RFM_Monetary AS RFM_Combined
FROM (
    SELECT CustomerID,
        NTILE(4) OVER (ORDER BY LastOrderDate) AS RFM_Recency,
        NTILE(4) OVER (ORDER BY OrderCount) AS RFM_Frequency,
        NTILE(4) OVER (ORDER BY TotalPrice) AS RFM_Monetary
    FROM clean_data
)
'''
```

NTILE Function of SQL : geeksforgeeks.org/ntile-function-in-sql-server/

```python
grouping = pd.read_sql(query, con)
```

```python
grouping
```

```python
#Turning this into a SQL table
grouping.to_sql('RFM_scores', con, if_exists = 'replace', index=False)
```

```python
#Let's see the customer with the lowest recency, frequency and monetary value
query = 'SELECT * FROM RFM_scores ORDER BY RFM_Recency, RFM_Frequency, RFM_Monetary'
pd.read_sql(query,con)
```

```python
#Let's check a few customers with a combined RFM score of 111

#First extract the list of customer ids with 111 RFM score
low_cx_1 = grouping[grouping['RFM_Combined'] == 111]
low_cx_2 = list(low_cx_1['CustomerID'])

#Now let's get the relavant records from the clean_data dataframe
lost_customers = clean_data[clean_data['CustomerID'].isin(low_cx_2)]
lost_customers
```

When we create a ranking by dividing each criteria of RFM into 4 bins, we create 4*4*4 = 64 distinctive groups. 

Since that much of groups are not practical, we can create custom groups as below:



### Logic Using For This:

* Best customers: R>=4, F>=4, M>=4
* Loyal: R>=3, F>=3, M>=3
* Potential Loyalist: R>=3, F>=1, M>=2
* Promising: R>=3, F>=1, M>=1
* Customers Needing Attention: R>=2, F>=2, M>=2
* At Risk: R>=1, F>=2, M>=2
* Hibernating: R>=1, F>=1, M>=2
* Else Lost

```python
#A function to apply the grouping
def rfm_level(rfm_group):
    if((rfm_group['RFM_Recency'] >= 4) and (rfm_group['RFM_Frequency'] >= 4) and (rfm_group['RFM_Monetary'] >= 4)):
        return "The Best Customers"
    elif((rfm_group['RFM_Recency'] >= 3) and (rfm_group['RFM_Frequency'] >= 3) and (rfm_group['RFM_Monetary'] >= 3)):
        return "Loyal Customers"
    elif((rfm_group['RFM_Recency'] >= 3) and (rfm_group['RFM_Frequency'] >= 1) and (rfm_group['RFM_Monetary'] >= 2)):
        return "Potential Loyalists"
    elif((rfm_group['RFM_Recency'] >= 3) and (rfm_group['RFM_Frequency'] >= 1) and (rfm_group['RFM_Monetary'] >= 1)):
        return "Promising Customers"
    elif((rfm_group['RFM_Recency'] >= 2) and (rfm_group['RFM_Frequency'] >= 2) and (rfm_group['RFM_Monetary'] >= 2)):
        return "Customers Needing Attention"
    elif((rfm_group['RFM_Recency'] >= 1) and (rfm_group['RFM_Frequency'] >= 2) and (rfm_group['RFM_Monetary'] >= 2)):
        return "At Risk"
    elif((rfm_group['RFM_Recency'] >= 1) and (rfm_group['RFM_Frequency'] >= 1) and (rfm_group['RFM_Monetary'] >= 2)):
        return "Hibernating"
    else:
        return "Lost"
```

```python
#Applying the function and creating a new variable
grouping['RFM_Level'] = grouping.apply(rfm_level, axis=1)
grouping
```

```python
#Let's export this to the SQL table
grouping.to_sql('RFM_scores', con, if_exists = 'replace', index=False)
```

```python
#Checking if exported is successful
pd.read_sql('SELECT * FROM RFM_scores LIMIT 10',con)
```

```python
grouping['RFM_Level'].value_counts()
```

## Creating a Tree Map for The Bins

```python
#Converting number of customers in each bin to a dataframe
rfm_bin_counts = grouping['RFM_Level'].value_counts().reset_index()
rfm_bin_counts
```

```python
#Creating the tree map using squarify
import squarify
fig = plt.gcf()
ax = fig.add_subplot()
fig.set_size_inches(13, 7)
squarify.plot(sizes=rfm_bin_counts['count'], 
              label=['Loyal Customers',
                     'Lost',
                     'Customers Needing Attention',
                     'Potential Loyalists',
                     'At Risk', 
                     'The Best Customers', 
                     'Promising Customers',
                     'Hibernating'], alpha=0.7)
plt.title("RFM Segments",fontsize=20)
plt.axis('off')
plt.show()
```
